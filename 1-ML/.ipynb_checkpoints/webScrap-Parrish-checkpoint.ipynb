```python
import requests
import urllib.request
import pandas as pd

resp = requests.get("https://www.tandfonline.com/doi/full/10.1080/23311983.2016.1236436")
html_str = resp.text
```

```python
html_str
```

```python
from bs4 import BeautifulSoup

document = BeautifulSoup(html_str, "xml")
document.prettify()
```

```python
match = document.title.text
match
```

```python
abstract = document.find('div', class_= 'abstractSection abstractInFull')
abstract
```

```python
abstract= document.find_all('p')
print (len(abstract))
for a in range(0,len(abstract)):
               print(abstract[a].text)
```
