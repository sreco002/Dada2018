```python
import requests

resp = requests.get("https://wepa.unima.org/en/theatre-of-objects/")
html_str = resp.text
```

```python
html_str
```

```python
from bs4 import BeautifulSoup

document = BeautifulSoup(html_str, "html.parser")
type(document)
```

```python
p_tags = document.find_all('p')
type(p_tags)
```

```python
[tag.string for tag in p_tags]
```

```python
h1_tags = document.find_all('h1')
type(h1_tags)
```

```python
[tag.string for tag in h1_tags]
```

```python
h2_tags = document.find_all('h2')
type(h2_tags)
```

```python
[tag.string for tag in h2_tags]
```
