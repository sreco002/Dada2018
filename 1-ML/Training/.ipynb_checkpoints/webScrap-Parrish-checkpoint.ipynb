```python
import requests
import urllib.request
import pandas as pd

resp = requests.get("https://www.tandfonline.com/doi/full/10.1080/23311983.2016.1236436")
html_str = resp.text
#html_str
```

```python
from bs4 import BeautifulSoup

document = BeautifulSoup(html_str, "lxml")
document.prettify()
```

```python
match = document.title.text
match
```

```python
abstract = document.find('div', class_= 'abstractSection abstractInFull')
abstract
```

```python
abstract= document.find_all('p')
#print (len(abstract))
#open a new  output file 'file.txt' . write in all the texts included in the html, close the file
output = open("file.txt", "w")
for a in range(3,len(abstract)-3):
        output.write(abstract[a].text)
output.close()
```

```python
with open ('file.txt',"r") as f:
   for line in f:
    print( line, end="")
    
```
